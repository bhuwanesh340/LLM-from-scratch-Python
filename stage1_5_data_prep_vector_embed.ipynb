### Creating Token Embeddings

```python
import torch
print("Pytorch version:", torch.__version__)
```

```python
input_ids = torch.tensor([2,3,5,1])
```

For example we have vocab of size 6 (QUICK FOX IS IN THE HOUSE) (instead of 50,257 in BPE tokenizer) 

Suppose we want to create embeddings of size 3 (in GPT-3 embedding size is 12,288 dimensions)

Using vocab_size and output_dim, we can instantiate an embedding layer in Pytorch

```python
vocab_size = 6
output_dm = 3

torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size, output_dm)
```

```python
print("Embedding layer weights:\n", embedding_layer.weight)
```

This initializes the weights in a random manner

We can see that the weight matrix of the embedding layer contains small, random values.

These values are optimized during LLM training as part of the LLM optimization itself.

there is one row for each possible tokens in the vocabulary & one column for each of the dimensions.

```python
print(embedding_layer(torch.tensor(2)))
```

This is a lookup value for row 3 from weight matrix via token ID

Let's fetch the vector for IDs (2,3,5,1) from the entire vocab

```python
print(embedding_layer(input_ids))
```
