### Byte Pair Encoding (BPE)

The BPE tokenizer covered below was used to train LLMs such as GPT-2, GPT-3 and original model used in GPT

```python
import tiktoken
import importlib
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

This tokenizer is similar to SimpleTokenizerV2 we have created

```python
text = (
    "Hello, do you like tea? <|endoftext|> In the sunlit terraces "
    "of someunknownPlace"
)

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})

print(integers)
```

We con convert the token IDs back into text using the decode method

```python
strings = tokenizer.decode(integers)

print(strings)
```
